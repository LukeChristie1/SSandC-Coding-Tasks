```python
import pandas as pd
import os
file_path = r'C:\Users\phlc\OneDrive - Loughborough University\SS&C tasks\anomaly_detetcion_dataset.csv'  
df = pd.read_csv(file_path)
columns_to_analyze = ['x1', 'x2', 'x4']
for column in columns_to_analyze:
        if column in df.columns:
            most_frequent = df[column].value_counts().nlargest(1)
            print(f"Most frequently occurring value in '{column}':")
            print(most_frequent)
        else:
            print(f"Column '{column}' does not exist in the DataFrame.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

file_path = r'C:\Users\phlc\OneDrive - Loughborough University\SS&C tasks\anomaly_detetcion_dataset.csv'
df = pd.read_csv(file_path)  

columns_to_standardize = ['x1']  

for column in columns_to_standardize:
    plt.figure(figsize=(10,6))
    sns.histplot(df[column], kde=True) # Kernel Density Estimate(smoothing of histogram)
    plt.title(f'Original Distribution: {column}', fontsize=16)
    plt.xlabel(column, fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.tight_layout()
    plt.show()

    scaler = StandardScaler()
    df[column] = scaler.fit_transform(df[[column]])

    plt.figure(figsize=(10,6))
    sns.histplot(df[column], kde=True)
    plt.title(f'Standardized Distribution: {column}', fontsize=16)
    plt.xlabel(f'Standardized {column}', fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

file_path = r'C:\Users\phlc\OneDrive - Loughborough University\SS&C tasks\anomaly_detetcion_dataset.csv'
df = pd.read_csv(file_path)  

columns_to_standardize = ['x6']  

for column in columns_to_standardize:
    plt.figure(figsize=(10,6))
    sns.histplot(df[column], kde=True)
    plt.title(f'Original Distribution: {column}', fontsize=16)
    plt.xlabel(column, fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.tight_layout()
    plt.show()
    
    log_data = np.log(df[column])
    

    plt.figure(figsize=(10,6))
    sns.histplot(log_data, kde=True)
    plt.title(f'Standardized Distribution: {column}', fontsize=16)
    plt.xlabel(f'Standardized {column}', fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```
